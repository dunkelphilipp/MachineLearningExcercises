```python
import pandas as pd

# Read file
iowa_file_path = '/Users/philippdunkel/PycharmProjects/MachineLearningExcercises/melb_data.csv'
home_data = pd.read_csv(iowa_file_path)
```

Step 1: Specify Prediction Target as y

```python
# print the list of columns in the dataset to find the name of the prediction target
home_data.columns
```

```python
# selecting the target variable to the Sales Price
y = home_data.Price
```

Step 2: create x

```python
# Create the list of features below
feature_names = ['Landsize', 'YearBuilt', 'Bathroom']

# Select data corresponding to features in feature_names
X = home_data[feature_names]
```

```python
# Review data
# print description or statistics from X
print(X.describe())

# print the top few lines
print(X.head())
```

Step 3: Specify and fit Model

```python
from sklearn.tree import DecisionTreeRegressor
#specify the model. 
#For model reproducibility, set a numeric value for random_state when specifying the model
iowa_model = DecisionTreeRegressor(random_state = 1)

# Fit the model
iowa_model.fit(X, y)
```

Step 4: Make predictions

```python
predictions = iowa_model.predict(X)
print(predictions)
```
